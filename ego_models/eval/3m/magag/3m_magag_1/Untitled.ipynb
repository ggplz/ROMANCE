```python
import numpy as np
```

```python
data = np.loadtxt("./test_return.txt")
```

```python
data.shape
```

```python
y = np.mean(data, axis=1)
```

```python
y
```

```python
x = np.arange(0, 801, 4)
x.shape
```

```python
from scipy.interpolate import make_interp_spline
model=make_interp_spline(x, y)
xs = np.arange(0, 801)
ys = model(xs)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(xs, -ys)
plt.ylim(0, 21) 
plt.xlim(0, 801)
plt.show()
```
